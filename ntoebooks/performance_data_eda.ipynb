```python

# Create the histogram
plt.figure(figsize=(10, 8))  

# Histogram for CTR
plt.hist(performance_data['CTR'], bins=10, alpha=0.7, label='CTR')  # bins control the number of bars

# Histogram for ER 
plt.hist(performance_data['ER'], bins=10, alpha=0.7, label='ER')  # Uncomment for a stacked histogram

# Density plot for CTR 
sns.kdeplot(performance_data['CTR'], label='CTR')  # Consider using seaborn for density plots

# Customize the plot
plt.xlabel('Click-Through Rate (CTR) or Engagement Rate (ER)')
plt.ylabel('Frequency')
plt.title('Distribution of CTR ')
plt.legend()
plt.grid(True)
plt.show()
```
